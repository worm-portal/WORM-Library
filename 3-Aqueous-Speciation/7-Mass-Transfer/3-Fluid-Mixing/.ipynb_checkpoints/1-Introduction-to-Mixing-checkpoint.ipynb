Let's say we have two fluids with different compositions:

1) Fluid **Ca_rich** has 3 milimolal $Ca^{2+}$ and 1 µmolal $HCO_3^-$
2) Fluid **CO2_rich** has 10 µmolal $Ca^{2+}$ and 3 milimolal $HCO_3^-$

Both fluids are undersaturated with respect to calcite. If either fluid comes in contact with calcite mineral, there will be a thermodynamic drive for the calcite to dissolve until equilibrium has been established. How do we know this? We'll confirm it in a moment after a speciation calculation.

The main question of this demo is: **what happens when we mix these two fluids together?**

Will the mixed fluid still be undersaturated with respect to calcite? Or is calcite expected to precipitate? (The title of the demo kind of gives away the surprise, but let's just pretend we didn't see it.)

Let's perform a calculation to model what might happen when the fluids are mixed.

```python
import aqequil

ae=aqequil.AqEquil(db="wrm")
```

Take a look at the input file, `Ca_CO2_mix.csv`. It contains our two samples, Ca_rich and CO2_rich.

First, we'll speciate them.

Note the parameter `charge_balance_on` has been set to "H+" in the speciation function. We'll come back to that a little later.

```python
speciation = ae.speciate(input_filename="Ca_CO2_mix.csv",
                         charge_balance_on="H+",
                         )
```

We can confirm that our total Ca+2 and HCO3- after speciation matches what is in the input file using the `lookup` function.

```python
speciation.lookup(["pH", "Ca+2_total", "HCO3-_total"])
```

Note the pH of the samples in the table above. Do these pH values match the input file? The samples are assigned pH 7 in the input file, and now the samples are quite different: Ca_rich is alkaline and CO2_rich is acidic. What going on here?

Recall that the parameter `charge_balance_on` was set to "H+" in the speciation step. This allows the pH to change to ensure that the total charge of dissolved cations and anions is balanced by the charge of protons. In the case of the "Ca_rich" sample, there is an overabundance of positive charge from dissolved calcium ions that cannot be balanced by the amount of dissolved bicarbonate. Therefore, the number of protons must decrease to achieve charge balance, and the pH increases. It is a similar story with the "CO2_rich" sample, except that pH must decrease to provide more protons to balance out the negative charge of an overabundance of bicarbonate.

Okay, moving on - it was stated earlier that both samples are undersaturated with respect to calcite. Let's confirm this by plotting mineral saturation states in each sample. A negative affinity indicates that the fluid is undersaturated with respect to a mineral:

```python
speciation.plot_mineral_saturation("Ca_rich")
```

```python
speciation.plot_mineral_saturation("CO2_rich")
```

Calcite has a negative affinity in each sample and is therefore undersaturated.

### Mixing Ca_rich and CO2_rich fluids 

Our samples can now be mixed now that they have been speciated.

Let's define the parameters of our mixing calculation. We can specify which fluid we want to mix with all other fluids in our input file. Since we only have two fluids, "Ca_rich" and "CO2_rich", let's choose "CO2_rich" for our `Mixing_Fluid`. This tells the code to mix "CO2_rich" with every other fluid in our input file, including itself.

(Hypothetically, if you had 100 fluids in your input file and selected "CO2_rich", then the CO2_rich fluid would be mixed with itself and all 99 other fluids.)

```python
Mix = aqequil.Mixing_Fluid(speciation=speciation,
                           sample_name="CO2_rich")
```

Then we have the option to specify other aspects of our reaction. This is a mass transfer style reaction, so we use the same `Prepare_Reaction` function that we would use to calculate a water-rock reaction.

The `mineral_suppression_option` lets us control whether minerals are allowed to form during our mixing calculation. Let's keep it at the default value of "None" to allow minerals to form.

The `xi_print_int` parameter determines the interval at which results must be reported as the reaction proceeds and the fluids are mixed. Recall that reaction progress in mass transfer calculations is typically represented on a scale of 0 to 1 by the variable Xi. Xi=0 is just before the fluids are mixed. Xi=1 is when the fluids are fully mixed and the reaction is complete. By setting `xi_print_int=0.1`, we are telling the code to report what's going on at Xi=0.1, 0.2, 0.3... and so on. This is *in addition* to any other Xi values that the code decides to report. Therefore, lowering the value of `xi_print_int` will produce longer tables of results and smoother figures but will take slightly longer to finish calculating.

```python
r = aqequil.Prepare_Reaction(reactants=[Mix],
                             mineral_suppression_option="None",
                             xi_print_int=0.1)
```

`xi_print_int` is just one of many options available in the `Prepare_Reaction` function. Remember that you can use `AqEquil.Prepare_Reaction?` in a blank cell to see what parameters are available, their defaults, and brief descriptions of what those parameters do. Remove the # in the cell below and execute it to see for yourself.

```python
#aqequil.Prepare_Reaction?
```

Now we can react the two fluids in a mixing calculation:

```python
speciation_mix = aqequil.react(speciation, r)
```

Now we choose one of our samples to examine the results of the mixing. Let's choose Fluid B.

*Why not choose to look at the results for CO2_rich?* Recall that we chose CO2_rich as our mixing fluid. That means CO2_rich is mixed with all samples, including itself. Therefore, choosing "CO2_rich" in the cell below would not be very interesting, because it would just show what happens when CO2_rich is mixed with itself: nothing!

```python
m = speciation_mix.mt("Ca_rich")
```

Now we can visualize results immediately with built-in plotting functions.

Many of these plots have the variable Xi on the x-axis. Recall that Xi represents reaction progress, or extent of mixing.

Here is how the pH of Ca_rich is predicted to change. It starts around pH 11.7 and gradually drops as more CO2_rich fluid is titrated in and allowed to equilibrate, bit by bit, until fully mixed.

```python
m.plot_pH(x_type="xi")
```

We can use the usual built-in plot types for mixing calculations just like any other mass transfer calculation. Check out some of the other mass transfer demos in the WORM Library if you are unfamiliar with what is possible.

For example, plotting how the concentrations of solutes in Ca_rich change as CO2_rich is gradually mixed in:

```python
m.plot_aqueous_species(x_type="xi",
                       y_type="molality",
                       charge_sign_at_end=True,
                       plot_species=["HCO3-", "CO2", "CO3-2",
                                     "Ca+2", "CaHCO3+", "CaCO3"])
```

When we set up the mixing calculation earlier, we allowed minerals to precipitate in our mixing calculation. We can use the `plot_product_minerals` function to see which minerals are predicted to precipitate (and how much) if allowed to reach equilibrium with the fluid mixture.

```python
m.plot_product_minerals(x_type="xi", y_type="mole", log_y=False)
```

Recall that our two fluids are undersaturated with respect to calcite. However, when they are mixed, the resulting mixture is supersaturated. This causes calcite to precipitate to achieve equilibrium with the fluid mixture.

How can two undersaturated fluids mix to create a supersaturated fluid? Take a look at the reaction to form calcite, $\text{CaCO}_3$, from dissolved ions:

$$
\text{Ca}^{2+} + \text{HCO}_3^- = \text{CaCO}_3 + \text{H}^+
$$

Each of fluid in this demo is rich in either calcium or bicarbonate ions, but not both. Separately, the fluids lack the necessary concentrations of both solutes to drive the formation of calcite.

This changes when the fluids are mixed. The resulting mixture has more than enough calcium and bicarbonate ions to precipitate calcite. Neat!
