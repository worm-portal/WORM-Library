By default, no charge balancing is performed and electrical imbalance is included in the speciation report. Alternatively, it is possible to balance charge on a chosen ion by including the following parameter in the speciate() function:

```python
charge_balance_on='H+', # species used for balance goes here
```

This allows the activity of the chosen species to shift to bring the total charge imbalance to 0 in each sample.

The sample input file used in this example, peru.csv, contains data from hot spring samples collected from the Peruvian Andes ([Newell & Scott 2020](https://doi.org/10.26022/IEDA/111569))

````python
import aqequil
ae = aqequil.AqEquil(db="wrm")
````

````python
# perform a speciation calculation with the default WORM database 'wrm'
speciation = ae.speciate(input_filename="peru.csv",
                         exclude=["Year", "Area"],
                         delete_generated_folders=True,
                         charge_balance_on='H+')
````

### Confirm charge balance

````python
speciation.lookup(['sigma(mz) cations', 'sigma(mz) anions', 'charge imbalance % of total charge'])
````

### Comparing unbalanced (input) and balanced speciations

It is possible to compare charge-balanced and non-charge-balanced speciation results. For pH, this can be done easily by comparing pH_(input) to pH, where the latter is has been adjusted for charge balance:

````python
speciation.barplot(["pH_(input)", "pH"], plot_width=7)
````

When balancing on other chemical species, plotting might get tricky. This is because plots don't accept variables with different units on the y-axis. For instance, if you balance on Na$^+$, the speciation will return a log activity of Na$^+$, while you might've entered Na$^+$ in your sample spreadsheet as molality or ppm. These cannot be plotted alongside each other because they have different units.

One solution is to run two speciations, one unbalanced and one balanced on Na$^+$, and then compare the speciated log activities of Na$^+$.

````python
# unbalanced speciation
unbalanced_speciation = ae.speciate(input_filename="peru.csv",
                         exclude=["Year", "Area"],
                         delete_generated_folders=True)

# speciation balanced on Na+
balanced_speciation = ae.speciate(input_filename="peru.csv",
                         exclude=["Year", "Area"],
                         delete_generated_folders=True,
                         charge_balance_on='Na+')
````

````python
# This cell takes the balanced and unbalanced
# speciation reports and joins them side-by-side.
# Column names from the unbalanced report end in '_(unbalanced)'
# Column names from the balanced report end in '_(balanced)'

import pandas as pd
import copy

excluded_cols = [("Year", ""), ("Area", "")] # metadata headers and subheaders
necessary_cols = [("pH", "pH"), ("Temperature", "degC")]

input_cols = excluded_cols+necessary_cols+[col for col in unbalanced_speciation.report if "_(input)" in str(col)]
input_df = unbalanced_speciation.report[input_cols]
output_cols_unbalanced = [col for col in unbalanced_speciation.report.columns if col not in input_cols]
output_df_unbalanced = unbalanced_speciation.report[output_cols_unbalanced]
output_df_unbalanced.columns = [(col[0]+"_(unbalanced)", col[1]) for col in output_cols_unbalanced]
output_cols_balanced = [col for col in balanced_speciation.report.columns if col not in input_cols]
output_df_balanced = balanced_speciation.report[output_cols_balanced]
output_df_balanced.columns = [(col[0]+"_(balanced)", col[1]) for col in output_cols_balanced]
output_df_all = pd.concat([input_df, output_df_unbalanced, output_df_balanced], axis=1)
cbal_comparison = copy.deepcopy(unbalanced_speciation)
cbal_comparison.report = output_df_all
````

````python
# compare samples before and after balancing on Na+. Display only the first five rows with .head()
cbal_comparison.lookup(["Na+_(unbalanced)", "Na+_(balanced)"]).head()
````

````python
# using a barplot to compare samples before and after balancing on Na+
cbal_comparison.barplot(['Na+_(unbalanced)', 'Na+_(balanced)'], plot_width=7, show_missing=True)
````
