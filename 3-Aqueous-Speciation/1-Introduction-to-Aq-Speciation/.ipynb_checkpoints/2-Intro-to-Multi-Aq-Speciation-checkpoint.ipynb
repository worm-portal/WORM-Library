# Multi-Sample speciation

This notebook demonstrates that multiple fluid samples can be speciated inside of a notebook. The results of the multi-sample speciation can be plotted in a variety of ways.

First, check out leong2021.csv, included with this notebook in the WORM-Library. This CSV contains sample data selected after [Leong *et al.* 2021](https://doi.org/10.1029/2020JB020756). These water samples were collected from natural springs in Oman ranging pH 7-12. The hyperalkaline fluids in this dataset were the result of serpentinization, hydrous alteration of ultramafic rocks.

Next, run this notebook. Each sample in leong2021.csv will be speciated and the results plotted. Compare the plots to Figure 3 of [Leong *et al.* 2021](https://doi.org/10.1029/2020JB020756) (figures are embedded in this notebook).

```python
# import the Python packages this notebook needs
import AqEquil
import matplotlib.pyplot as plt
```

```python
# tell the notebook to make the figures larger
plt.rcParams['figure.figsize'] = [10, 5] 
```

```python
# get the aqueous speciation tool ready
ae = AqEquil.AqEquil()
```

```python
# perform a multi-sample speciation calculation
speciation = ae.speciate(input_filename="leong2021.csv",
                         exclude=["Year", "Area"], # exclude metadata columns
                         report_filename="leong_report.csv") # create a CSV of results 
```

Now that the samples are speciated, let's get a general idea of the pH and temperature range in this sample set.

```python
# create a scatterplot of pH and Temperature
speciation.scatterplot('pH', 'Temperature')
```

Samples are all below 40 Â°C and between pH 7 and 12. Many are between pH 11 to 12.

Now let's examine the results of the multi-sample speciation, starting with SiO$_{2(aq)}$, and see how it is possible to recreate Figure 3 of Leong *et al.* 2021.

---
# Silicon

<img src="https://gitlab.com/worm1/worm-figures/-/raw/master/WORKSHOP/si.png" width=400>

From Figure 3 of Leong *et al.* 2021

```python
# visualize basis species contributions to aqueous species by mass
speciation.plot_mass_contribution("SiO2", width=1, sort_by='pH',
    sort_y_by=["Mg(HSiO3)+", "Ca(HSiO3)+", "NaHSiO3", "HSiO3-", "SiO2", "Other"]) # bar order
```

---
# Magnesium

<img src="https://gitlab.com/worm1/worm-figures/-/raw/master/WORKSHOP/mg.png" width=400>

From Figure 3 of Leong *et al.* 2021

```python
speciation.plot_mass_contribution("Mg+2", width=1, sort_by='pH',
    sort_y_by=['Mg(HCO3)+', 'Mg(CO3)', 'Mg+2', 'MgOH+', 'MgSO4', 'Other'])
```

---
# Calcium

<img src="https://gitlab.com/worm1/worm-figures/-/raw/master/WORKSHOP/ca.png" width=400>

From Figure 3 of Leong *et al.* 2021

```python
speciation.plot_mass_contribution("Ca+2", width=1, sort_by='pH',
    sort_y_by=['Ca(HCO3)+', 'Ca(CO3)', 'Ca+2', 'CaSO4', 'CaOH+', 'Other'])
```

---
# Dissolved Inorganic Carbon

<img src="https://gitlab.com/worm1/worm-figures/-/raw/master/WORKSHOP/dic.png" width=400>

From Figure 3 of Leong *et al.* 2021

```python
speciation.plot_mass_contribution("HCO3-", sort_by="pH", width=1,
    sort_y_by=['Mg(HCO3)+', 'Mg(CO3)', 'HCO3-', 'CO2',
               'CO3-2','Ca(HCO3)+', 'Ca(CO3)', 'Other'])
```

A variety of other plotting options are available. For example, here is a multi-series scatterplot of several carbonate species:

```python
# Plot multiple dependent variables by providing scatterplot() with a list.
speciation.scatterplot("pH", ["HCO3-", "CO2", "Mg(HCO3)+", "CO3-2", "Ca(CO3)"])
```

It is also possible to look up species directly in a table with lookup():

```python
speciation.lookup(["HCO3-", "CO2", "Mg(HCO3)+", "CO3-2", "Ca(CO3)"])
```

For a brief overview of other plotting and lookup options refer to Multi-Sample-Speciation.ipynb in the WORM-Library.
