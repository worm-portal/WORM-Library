# Figure S2) Parameter sensitivity for both cold and hot fluid reservoirs.
Here we create Figure S2 in Trinh+ (2026), which shows a parameter sensitivity analysis. This notebook is like `Make-Figure-4.ipynb`, but here we include the cold reservoir sensitivity too.

Once again, we import our libraries and define helper functions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
def getDF(filename):
    '''Obtain DataFrame of energy supply and solute concentration data for a 
    given mixture. The "filename" is a list of mixture objects, and "i_nom" is the 
    index of the desired mixture.'''
    # open file of fluid mixing
    with open (filename, 'rb') as file:
        mixture = pickle.load(file)
    
    # obtain energy supply DataFrame
    df_E = mixture.plot_energy(y_type = "E",
            y_units = "cal",
            x_type='temperature',
            show_zero_line = True,
            species = ["CO2", "H2", "CH4", "H2O"],
            stoich = [-1, -4, 1, 2],
            df_out = True,)
    df_E = df_E[1]

    # translate Xi to fluid mass ratio
    df1 = df_E[df_E['Xi']<=1] # values where vent fluid < seawater mass
    tmp1 = 1/np.array(df1['Xi'].tolist()[1:])
    tmp1C = df1['Temp(C)'].tolist()[1:]
    
    df2 = df_E[df_E['Xi']>=1] # values where vent fluid > seawater mass
    tmp2 = 2-np.array(df2['Xi'].tolist())
    tmp2C = df2['Temp(C)']

    mass_ratios = tmp1.tolist() + tmp2.tolist() # kg of seawater added to vent fluid

    # obtain aqueous species DataFrame
    sp = ['CO2', 'CH4', 'H2', 'HCO3-', 'H+']
    df_aq = mixture.aq_distribution_logmolal[sp]

    # compute minimum energy supply
    Alim = 10.6e3 # minimum energy supply for methanogens (J / mol CH4) (Hoehler+, 2001)
    J2cal = 0.2390 # calories in a joule
    H2_molality = 10**df_aq['H2'] # concentration of H2 (molal)
    Emin = Alim*J2cal*H2_molality/4

    # specify columns of interest
    tmp_data = {'Xi': df_E['Xi'],
              'Mass ratio': mass_ratios,
              'Temp(C)': df_E['Temp(C)'],
              'Press(bars)': df_E['Press(bars)'],
              'pH': df_E['pH'],
              'log fO2': df_E['log fO2'],
              'CO2': df_aq['CO2'],
              'CH4': df_aq['CH4'], 
              'H2': df_aq['H2'],
              'HCO3-': df_aq['HCO3-'],
              'H+': df_aq['H+'],
              'E (cal/kg fluid)': df_E['E, cal/kg fluid'],
              'Emin (cal/kg fluid)': Emin
             }
    df = pd.DataFrame(tmp_data)
    return df

def makeFigure4(filenames, include_labels=True):
    '''Plot nominal model and individual parameter sensitivity.'''
    # get DataFrames for each model
    dfarr = []
    for i in range(len(filenames)):
        dfarr.append(getDF(filenames[i]))
    dfN, df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15, df16, df17, df18 = dfarr

    # helpful misc.
    Tmax = 500
    eng = 'E (cal/kg fluid)'
    
    # create figure skeleton
    fsize = (8, 10)
    fs = 8 # font size
    fig, axes = plt.subplots(nrows=4, ncols=2, figsize=fsize)
    ax0, ax1, ax2, ax3, ax4, ax5, ax6, ax7 = axes.flatten()

    # plot styling
    lw_nom = 2
    lw_ps = 1
    cless = 'tab:blue'
    cmore = 'tab:red'

    # plot Figure 4a (cold res - pH)
    ax0.plot(dfN['Temp(C)'], dfN[eng], color='black', lw=lw_nom)
    ax0.plot(df1['Temp(C)'], df1[eng], color=cless, lw=lw_ps)
    ax0.plot(df2['Temp(C)'], df2[eng], color=cmore, lw=lw_ps)
    ax0.plot([0, 500], [0, 0], 'k--')
    ax0.set_xlim([0, Tmax])
    ax0.set_xticklabels([])
    ax0.tick_params(top=True, axis='both', labelsize=8)
    ax0.axvspan(122, Tmax, facecolor='black', alpha=0.1)

    # plot Figure 4b (hot res - pH)
    ax1.plot(dfN['Temp(C)'], dfN[eng], color='black', lw=lw_nom)
    ax1.plot(df9['Temp(C)'], df9[eng], color=cless, lw=lw_ps)
    ax1.plot(df10['Temp(C)'], df10[eng], color=cmore, lw=lw_ps)
    ax1.plot([0, 500], [0, 0], 'k--')
    ax1.set_xlim([0, Tmax])
    ax1.set_xticklabels([])
    ax1.tick_params(top=True, axis='both', labelsize=8)
    ax1.axvspan(122, Tmax, facecolor='black', alpha=0.1)

    # plot Figure 4c (cold res - total C)
    ax2.plot(dfN['Temp(C)'], dfN[eng], color='black', lw=lw_nom)
    ax2.plot(df3['Temp(C)'], df3[eng], color=cless, lw=lw_ps)
    ax2.plot(df4['Temp(C)'], df4[eng], color=cmore, lw=lw_ps)
    ax2.plot([0, 500], [0, 0], 'k--')
    ax2.set_xlim([0, Tmax])
    ax2.set_xticklabels([])
    ax2.tick_params(top=True, axis='both', labelsize=8)
    ax2.axvspan(122, Tmax, facecolor='black', alpha=0.1)

    # plot Figure 4d (hot res - total C)
    ax3.plot(dfN['Temp(C)'], dfN[eng], color='black', lw=lw_nom)
    ax3.plot(df11['Temp(C)'], df11[eng], color=cless, lw=lw_ps)
    ax3.plot(df12['Temp(C)'], df12[eng], color=cmore, lw=lw_ps)
    ax3.plot([0, 500], [0, 0], 'k--')
    ax3.set_xlim([0, Tmax])
    ax3.set_xticklabels([])
    ax3.tick_params(top=True, axis='both', labelsize=8)
    ax3.axvspan(122, Tmax, facecolor='black', alpha=0.1)

    # plot Figure 4e (cold res - T)
    ax4.plot(dfN['Temp(C)'], dfN[eng], color='black', lw=lw_nom)
    ax4.plot(df5['Temp(C)'], df5[eng], color=cmore, lw=lw_ps)
    ax4.plot(df6['Temp(C)'], df6[eng], color=cmore, lw=lw_ps)
    ax4.plot([0, 500], [0, 0], 'k--')
    ax4.set_xlim([0, Tmax])
    ax4.set_xticklabels([])
    ax4.tick_params(top=True, axis='both', labelsize=8)
    ax4.axvspan(122, Tmax, facecolor='black', alpha=0.1)

    # plot Figure 4f (hot res - T)
    ax5.plot(dfN['Temp(C)'], dfN[eng], color='black', lw=lw_nom)
    ax5.plot(df13['Temp(C)'], df13[eng], color=cless, lw=lw_ps)
    ax5.plot(df14['Temp(C)'], df14[eng], color=cless, lw=lw_ps)
    ax5.plot(df15['Temp(C)'], df15[eng], color=cless, lw=lw_ps)
    ax5.plot(df16['Temp(C)'], df16[eng], color=cless, lw=lw_ps)
    ax5.plot([0, 500], [0, 0], 'k--')
    ax5.set_xlim([0, Tmax])
    ax5.set_xticklabels([])
    ax5.tick_params(top=True, axis='both', labelsize=8)
    ax5.axvspan(122, Tmax, facecolor='black', alpha=0.1)

    # plot Figure 4g (cold res - log fO2)
    ax6.plot(dfN['Temp(C)'], dfN[eng], color='black', lw=lw_nom)
    ax6.plot(df7['Temp(C)'], df7[eng], color=cless, lw=lw_ps)
    ax6.plot(df8['Temp(C)'], df8[eng], color=cmore, lw=lw_ps)
    ax6.plot([0, 500], [0, 0], 'k--')
    ax6.set_xlim([0, Tmax])
    ax6.tick_params(top=True, axis='both', labelsize=8)
    ax6.axvspan(122, Tmax, facecolor='black', alpha=0.1)

    # plot Figure 4h (hot res - log fO2)
    ax7.plot(dfN['Temp(C)'], dfN[eng], color='black', lw=lw_nom)
    ax7.plot(df17['Temp(C)'], df17[eng], color=cless, lw=lw_ps)
    ax7.plot(df18['Temp(C)'], df18[eng], color=cmore, lw=lw_ps)
    ax7.plot([0, 500], [0, 0], 'k--')
    ax7.set_xlim([0, Tmax])
    ax7.tick_params(top=True, axis='both', labelsize=8)
    ax7.axvspan(122, Tmax, facecolor='black', alpha=0.1)

    # add labels and tick values (or not)
    if include_labels:
        for ax in [ax0, ax2, ax4, ax6]:
            ax.set_ylabel('Energy supply\n(cal / kg of fluid)', fontsize=fs)
        ax6.set_xlabel(r'$T_{mix}$ (C)')
        ax7.set_xlabel(r'$T_{mix}$ (C)')
    else:
        for ax in axes.flatten():
            ax.set_yticklabels([])
            ax.set_xticklabels([])
        plt.savefig('Figure S2 - raw.svg')
```

# Model index map
This may be useful for identifying individual models:

- 0 = nominal model
  
Cold resevoir sensitivity
- 1 = neutral pH - 2
- 2 = neutral pH + 2
- 3 = $10^{-3}$ molal C
- 4 = $10^{-1}$ molal C
- 5, 6 = 10, 25 deg C
- 7 = PPM - 3
- 8 = PPM + 3

Hot reservoir sensitivity
- 9 = neutral pH - 2
- 10 = neutral pH + 2
- 11 = $10^{-3}$ molal C
- 12 = $10^{-1}$ molal C
- 13-16 = 300, 350, 400, 450 deg C
- 17 = PPM - 3
- 18 = PPM + 3

Now let's plot everything.

```python
mnames = []
for i in range(19):
    mnames.append('ps_output_' + str(i) + '.pkl')
makeFigure4(mnames, include_labels=True)
```

Interestingly, the parameter sensitivity test for the cold fluid hardly changes our results! It's really all about the variations of T and fO2 in the hot reservoir.
