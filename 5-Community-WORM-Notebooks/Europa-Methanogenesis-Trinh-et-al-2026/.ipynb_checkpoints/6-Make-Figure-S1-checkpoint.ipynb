# Figure S1) Plot equilibrium constant for CO2/HCO3- reaction.
Here we create Figure S1 in Trinh+ (2026), which shows the equilibrium constant of the following reaction:

$CO_{2(aq)} + H_2O \rightleftharpoons HCO_{3(aq)}^- + H^+_{(aq)}$

Once again, we import our libraries and define helper functions.

```python
import numpy as np
import aqequil
import matplotlib.pyplot as plt
import pychnosz
```

Calculate the equilibrium constant of the bicarbonate dissolution reaction.

```python
# define T-P arrays
T_arr = np.linspace(0.01, 550, 100)
P_arr = 1000 * np.ones(100)

# calculate logK at given T-P conditions
data_CO2 = pychnosz.subcrt(
        species = ['CO2', 'H2O', 'HCO3-', 'H+'],
        state = ['aq', 'liq', 'aq', 'aq'],
        coeff = [-1, -1, 1, 1],
        T = T_arr,
        P = P_arr,
        property="logK",
        messages=False,
        show=True,
        ).out
```

Now plot the data.

```python
plt.figure()
plt.plot(T_arr, data_CO2.logK)
plt.xlim([0, 500])
plt.xlabel(r'Temperature ($^o$C)')
plt.ylabel(r'$log_{10}K$')
plt.title(r'$CO_{2(aq)} + H_2O \rightleftharpoons HCO_{3(aq)}^- + H^+_{(aq)}$')
```
