## Figure 2) Oxidation State of Fluid Mixing
Here we create Figure 2 in Trinh+ (2026), which shows how the oxidation state of a fluid mixture varies as a function of mixed temperature.

First, we import AqEquil and the necessary libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib
import pickle
import pychnosz
import aqequil
%matplotlib inline

ae = aqequil.AqEquil(db='kev', 
                     suppress_redox=["C"],
                     exclude_organics=True,
                     )
```

Next, we calculate the equilibrium constants for the following reactions at T = 0.01â€“550 deg C and P = 1 kbar:

- $CO_{2(aq)} + 2 H_2O_{liq} \rightleftharpoons CH_{4(aq)} + 2 O_{2(aq)}$
  
- $3 FeS_{2(s)} + 2 Fe_3O_{4(s)} \rightleftharpoons 6 FeS_{(s)} + 2 O_{2(aq)}$

- $O_{2(g)} + 2 H_{2(g)} \rightleftharpoons 2 H_2O_{(liq)}$

- $O_{2(g)} \rightleftharpoons O_{2(liq)}$

These reactions represent the CO2/CH4 conversion, PPM redox buffer, the H2O dissociation reaction, and the conversion between gaseous and aqueous O2.


```python
# define parameter range to plot, initially
T_arr = np.linspace(0, 550, 101)
Pres = 1e3

# compute equilibrium constant for specified T-P, methanogenesis
data = pychnosz.subcrt(
        species = ['CO2', 'H2O', 'CH4', 'O2'],
        state = ['aq', 'liq', 'aq', 'aq'],
        coeff = [-0.5, -1, 0.5, 1],
        T = T_arr.tolist(),
        P = Pres,
        property="logK",
        messages=False,
        show=False,
        ).out

# PPM buffer
data_PPM = pychnosz.subcrt(
        species = ['pyrite', 'magnetite', 'pyrrhotite', 'O2'],
        state = ['cr', 'cr', 'cr', 'aq'],
        coeff = [-1.5, -0.5, 3, 1],
        T = T_arr.tolist(),
        P = Pres,
        property="logK",
        messages=False,
        show=False,
        ).out

# relationship between fH2 and fO2
data_H2O = pychnosz.subcrt(
        species = ['O2', 'H2', 'H2O'],
        state = ['gas', 'gas', 'liq'],
        coeff = [-1, -2, 2],
        T = T_arr.tolist(),
        P = Pres,
        property="logK",
        messages=False,
        show=False,
        ).out

# equilibrium between aqueous and gaseous O2
data_O2 = pychnosz.subcrt(
        species = ['O2', 'O2'],
        state = ['gas', 'aq'],
        coeff = [-1, 1],
        T = T_arr.tolist(),
        P = Pres,
        property="logK",
        messages=False,
        show=False,
        ).out

# calculate bounds for possible log fO2
logfO2_min_arr = [] # determined by max fH2
logfO2_max_arr = np.ones(len(T_arr))*np.log10(Pres)
for i in range(len(T_arr)):
    logfO2_min = -2*np.log10(Pres) - data_H2O['logK'][i]
    logfO2_min_arr.append(logfO2_min)
```

Now that we have our equilibrium constants, we can plot the our results! The figure below is saved as an svg file without axes labels to prepare for post-processing a publication-formatted figure. See Figure 2 in Trinh+ (2026) for the formatted and labelled version.

```python
# include axes labels?
lflag = True

# decide on T and aO2 of two fluids
p1, p2 = [2, -88], [500, -22.5] # (T, aO2)
psize = 50 # point size

# obtain plotting data
xdata = data['T']
ydata = data['logK']
ydata_PPM = data_PPM['logK']

# set plotting aesthetics
plt.rcParams.update({'font.size': 12})
plt.minorticks_on()

plt.rcParams['figure.figsize'] = [6, 5]
lw_buffer = 0.5 # line width for mineral redox buffer

# plot curves and regions based on the eq. constants from the last cell
ax = plt.gca()
ax.plot(xdata, ydata_PPM, lw=lw_buffer, c='black', linestyle='solid', label='PPM buffer')
ax.plot(xdata, ydata, 'k--', lw=2, label=r'$CO_2$ to $CH_4$ conversion')
ax.tick_params(axis='x', which='both', bottom=True, top=True)
ax.tick_params(axis='y', which='both', left=True, right=True)
ax.scatter(p1[0], p1[1], s=psize, c='tab:blue')
ax.scatter(p2[0], p2[1], s=psize, c='tab:red')
ax.tick_params(axis='both', 
               which='major', 
               top=True, 
               right=True,
               length=5,
               width=0.75
              )
ax.tick_params(axis='both', 
               which='minor', 
               top=True, 
               right=True,
               length=2,
               width=0.5
              )
plt.fill_between(T_arr, -100, logfO2_min_arr, color='black', alpha=1, label='Liquid water unstable')
ax.set_xlim([0, 550])
ax.set_ylim([-100, 0])
if lflag:
    ax.set_xlabel('Temperature (C)')
    ax.set_ylabel(r'$log_{10} a O_2$')
else:
    ax.set_xticklabels([])
    ax.set_yticklabels([])

# plot mixing path
mix_path = True
if mix_path:
    with open ('ps_output_0.pkl', 'rb') as file:
        mixture = pickle.load(file)
    df_E = mixture.plot_energy(y_type = "E",
            y_units = "cal",
            x_type='temperature',
            show_zero_line = True,
            species = ["CO2", "H2", "CH4", "H2O"],
            stoich = [-1, -4, 1, 2],
            df_out = True,)
    df_E = df_E[1]
    mix_T = df_E['Temp(C)'].tolist()
    df_aq = mixture.aq_distribution_logmolal
    logaO2 = df_aq['O2']
    ax.plot(mix_T, logaO2, color='tab:purple')

plt.savefig('Figure 2 - raw.svg')
plt.show()
```

The plotting elements are as follows: the CO2/CH4 conversion (black line, thick dashed), PPM buffer (black line, thin solid), H2O instability (solid black region), fluid mixture (purple line, solid), cold reservoir (blue circle), and hot reservoir (red circle). The temperature of the mixed fluid is based on the mass ratio between the cold and hot reservoirs.
