This notebook is used for generating a custom data0 file (i.e., a thermodynamic database) that is designed for the conditions that we are interested in. Specifically, we want to a data0 file that supports calculations at T = 0.01â€“500 deg C and P = 1 kbar. We also want to suppress redox reactions involving carbon, which allows for the chemical disequilibria that drives methanogenesis during fluid mixing.

All data0 files require a three-character file extension. For this study, we use extension "kev" to save all data files.

```python
import numpy as np
import aqequil
ae=aqequil.AqEquil(exclude_organics=True,
                   suppress_redox=["C"],
                   )
T_arr = np.linspace(0.01, 550, 8).tolist()
P_arr = [1000]*8
ae.create_data0(db="kev", 
                grid_temps=T_arr,
                grid_press=P_arr)
ae.runeqpt("kev")
```
