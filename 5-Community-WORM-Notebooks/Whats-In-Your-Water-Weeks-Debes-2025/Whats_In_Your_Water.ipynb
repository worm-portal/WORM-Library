## Introduction to Minerals

Minerals are the pure crystalline solid forms of chemical compounds with distinct physical and chemical properties. Minerals can be dissolved by water leading to increased aqueous (dissolved in water) concentrations of the mineral’s component elements (ions). You have probably observed this “water-rock reaction” on a much smaller scale in your kitchen when you make a glass of salt water. Common table salt (NaCl) is a mineral called halite.

Chemical analysis allows us to measure the concentration of dissolved ions (from minerals) and $H^{+}$ in water. However, it is important to note that once minerals are dissolved in water, their dissolved ionic parts can combine and form new compounds that may be different from the starting compounds. For example positive sodium ions, $Na^{+}$, from halite can combine with a negative sulfate ion, $SO_{4}^{2-}$, from anhydrite dissolving to form a new compound, $Na_{2} SO_{4}$. This “new” compound was not part of the starting minerals (halite or anhydrite) but has parts from each starting ingredient. The recombination of positive and negative ions from multiple initial compounds to form new compounds is called chemical speciation.

Let's examine how lead speciates and investigate the behavior of lead carbonate dissolution at varying pH values. To achieve this, we will use a water sample that serves as a proxy for the Flint River's chemistry and determine the equilibrium of the reaction below across varying pH values. 

Here is the generalized chemical reaction for this model: 
$PbCO_{3} (s) + H^{+} -> Pb^{2+} (aq) + HCO_{3}^{-} (aq)$

1. Import the Required Libraries.

```python
#Import the required libraries
import plotly.express as px
import pandas as pd
import aqequil
ae = aqequil.AqEquil(db="wrm")
```

2. Perform chemical speciation of the Flint River proxy chemistry at various pH values.
In the quotations next to report_filename, name your output file. Then run the cell.

```python
#Let's speciate our data across a range of pH!
speciation = ae.speciate(input_filename="Flint_River_Across_pH.csv", #Name your input file
                         aq_dist_type="molality",                    #Provides concentrations as molalities
                         charge_balance_on="Na+",                    #Ignore
                         report_filename= "output.csv")  #Name your output file
```

4. Set up a reaction between the water source and the lead carbonate mineral, cerussite.

```python
Cer = aqequil.Reactant(reactant_name="cerussite",
                        amount_remaining=0.00374) # moles in 1 g of cerussite

r = aqequil.Prepare_Reaction(reactants=[Cer])
```

```python
speciation = aqequil.react(speciation, r)
```

5. Organize the data produced from reacting lead carbonate (cerussite) with the water source.

```python
# Initialize empty lists to store results for each sample
pH_list = []              # Initial pH of each sample
pH_eq_list = []           # Equilibrium (final) pH after reaction modeling
conc_Pb_aq_list = []      # Aqueous lead concentration (in µg/kg or µg/L depending on context)
mass_Pb_solid_list = []   # Total mass of Pb in solid phase (e.g., as cerussite)
EPA_level = []            # EPA regulatory limit for Pb concentration in water (15 µg/L)

# Loop through each sample in the speciation report
for i, sample in enumerate(speciation.report.index):
    # Retrieve the model object for this sample
    m = speciation.mt(sample)

    # ----------------------------
    # Record the initial solution pH
    # ----------------------------
    pH = speciation.report["pH"].iloc[i, 0]
    pH_list.append(pH)

    # ----------------------------
    # Record the equilibrium pH from the model output
    # ----------------------------
    pH_eq = m.misc_params["pH"].iloc[-1]
    pH_eq_list.append(pH_eq)

    # ----------------------------
    # Extract aqueous Pb concentration from final model state
    # Convert from molal (mol/kg H2O) to µg/kg using atomic weight of Pb (207.2 g/mol)
    # ----------------------------
    conc_Pb_aq = m.dissolved_elements_molal["Pb"].iloc[-1]
    conc_Pb_aq_list.append(conc_Pb_aq * (207.2 * 1_000_000))

    # ----------------------------
    # Append constant EPA limit for Pb (15 µg/L)
    # ----------------------------
    EPA_level.append(15)

    # ----------------------------
    # Calculate the total solid-phase Pb (if cerussite formed)
    # Multiply moles of cerussite by molar mass of Pb (approximation)
    # ----------------------------
    mass_Pb_solid = 0
    if "cerussite" in m.moles_product_minerals.columns:
        mass_Pb_solid += m.moles_product_minerals["cerussite"].iloc[-1] * 207.2  # mol * g/mol = g

    mass_Pb_solid_list.append(mass_Pb_solid)
```

```python
#------------------
# Creating a dataframe that contains the data organized above
#------------------
df = pd.DataFrame({"initial pH":pH_list,
                   "pH after equilibration":pH_eq_list,
                   "Total dissolved Pb, ppb":conc_Pb_aq_list,
                   "Mass of Pb in scale, g":mass_Pb_solid_list,
                   "EPA Action Level, ppb":EPA_level},
                   index=speciation.report.index)
df
```

6. Visualize our results by plotting the equilibrium of cerussite as a function of pH.

```python
#Now let's plot our data!
fig = px.line(df, x="initial pH", y="Total dissolved Pb, ppb", template="simple_white") #Setting up the plot
fig.update_layout(yaxis_title="Total Dissolved Lead (ppb)")                             #Labelling the y-axis
fig.update_layout(xaxis_title="pH")                                                     #Labelling the x-axis
fig.show()                                                                              #Printing out the figure
```
